##2.2 NumPy Functions Part 1

the **np.arange function**, it returns an array with the specified start and stop and maybe intervals if selected also.
Remember in python the end is always excluded.

Using np.arange() function

The np.arange() function returns an array with evenly spaced elements as per the interval. The interval mentioned is half-opened i.e. start is included but stop is excluded.
It has the following paramaters:
start : start of interval range. By default start = 0
stop : end of interval range
step : step size of interval. By default step size = 1

```python

arr2 = np.arange(start= 0, stop= 10)
print(arr2)
arr3 = np.arange(start= 0, stop= 10, step=2)
print(arr3)
```

**np.linspace() function**

##2.6 Accessing Entries of a NumPy Array

Pandas is a library used to create datatypes and to manipulate datatypes.
A panda series is a 1D labeled array/list capable of holding data of any type.
The labels are collectibley called index
panda sereies can be imagined as a column on an excel spreadsheet.



```python
#create a list of medicine prices
med_price_list = [55,25,75,40,90]

#converting th med price list to an array
medprice= np.array(med_price_list )

#convert to a series
series_list = pd.Series(med_price_list)
series_arr =pd.Series(medprice)

#print to display
print(series_arr)
print(series_list)

#Change the index of a series
med_price_list_labeled= pd.Series(med_price_list, index = ['Omeprazole', 'Lexapro', 'Ibuprofen', 'Acetominaphen', 'Metformin'])
print(med_price_list_labeled)

#Updating prices for ***inflationnnnn**
med_price_list_updated=med_price_list_labeled + 2.5
print(med_price_list_updated)


#you can even compare the prices between the series
print(med_price_list_updated-med_price_list_labeled)

student = ['Mary', 'Peter', 'Susan', 'Toby', 'Anna']
df1 = pd.DataFrame(student, columns=['Student'])
df1

grades= ['B-','A+','B+','C','A-']
df2 = pd.DataFrame({'Student':student, 'Grade':grades})
df2


year= pd.Series([2012,2013,2014,2015,2016,2017,2018])
energy_consumption = pd.Series([2152, 2196, 2217, 2194, 2172, 2180, 2258])

df3 = pd.DataFrame({'Year':year, 'Energy_Consumption(Mtoe)':energy_consumption})
df3

```

##2.14 Pandas- Combining Data Frames


```python
#how to combine data frames (using concat, join, and merge)

data_cust=pd.DataFrame({'CustomerID': ['101','102','103','104'],
                        'category':['Medium','Medium','High', 'Low'],
                        'first_visit':['yes','no','yes', 'yes'],
                        'sales':[123,52,214,663]},index=[0,1,2,3])

data_cust_new= pd.DataFrame({"CustomerID":['101','103','104','105'],
                             'distance':[12,9,44,21],
                             'sales':[123,214,663,331]},index = [4,5,6,7])
print(data_cust)
print(data_cust_new)
```

```python
pd.concat([data_cust,data_cust_new], axis=0)# references rows
```

```python
pd.concat([data_cust,data_cust_new], axis=1)
```

```python
#merges using common values
pd.merge(data_cust, data_cust_new, how='outer', on='CustomerID')
```

```python
pd.merge(data_cust, data_cust_new, how='inner', on='CustomerID')
```

```python
pd.merge(data_cust, data_cust_new, how='right', on='CustomerID')
```

```python
data_quarters=pd.DataFrame({'Q1':[101,102,103],
                            'Q2':[201,202,203]},index=['I0','I1','I2'])

data_quarters_new=pd.DataFrame({'Q3':[301,302,303],
                                'Q4':[401,402,403]},index=['I0','I1','I3'])
print(data_quarters)
print(data_quarters_new)

data_quarters.join(data_quarters_new, how='right') #outer, inner, left and right work the same as merge
```
